```python
import pandas as pd
import pickle
```

```python
with open('sephora_data.pkl', 'r') as picklefile:
    sephora = pickle.load(picklefile)
```

```python
sephora = pd.DataFrame(sephora).transpose().reset_index()
sephora.head()
```

```python
sephora.sort_values(by='rating', ascending=False).head(10)
```

```python
sephora.rating.mean()
```

```python
c = 10
prior = 3
```

```python
def total_stars (x):
    return x.num_reviews * x.rating

def bayesian_estimate (x):
    return (c * prior + x.rating * x.num_reviews) / (c + x.num_reviews)
```

```python
sephora['total_stars'] = sephora.apply(total_stars, axis=1)
sephora['bayesian_est'] = sephora.apply(bayesian_estimate, axis=1)
```

```python
sephora.sort_values(by='bayesian_est', ascending=False).head(10)
```

```python
sephora.sort_values(by='rating', ascending=False).head(15)
```

```python
sephora['name'] = sephora.name.map(lambda x: x.encode('utf-8', 'replace'))
sephora['brand'] = sephora.brand.map(lambda x: x.encode('utf-8', 'replace'))
```

```python
sephora.sort_values(by='rating', ascending=False).head(15).to_csv('visualizations/sephora.csv', index=False)
```

```python
sephora.sort_values(by='bayesian_est', ascending=False).to_csv('visualizations/sephora-table.csv', index=False)
```
